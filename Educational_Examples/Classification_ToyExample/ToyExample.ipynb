First, we read in the well-known _Titanic_ dataset and display its first five rows.

```python
# load the `pandas` package
import pandas as pd

# read in the Titanic dataset
df = pd.read_csv('Titanic.csv')

# print first 5 rows of the Titanic dataset
df.head(5)
```

Next, we use the `seaborn` package to plot the correlation matrix of a few features of the Titanic dataset.

```python
# load the `seaborn` package
import seaborn as sns

# calculate correlation matrix
corr_matrix = df[['Age', 'Pclass', 'Fare', 'Survived']].corr()

# plot correlation matrix
sns.heatmap(corr_matrix, annot=True, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(0, 255, sep=48, n=256));
```
